## 18.3 Area Plot
Its basically a line chart with area colored under the graph

```python
import numpy as np
import pandas as pd

# import pyplot module for plotting
import matplotlib.pyplot as plt

# import warnings to ignore warning messages for now -
import warnings
warnings.filterwarnings("ignore")
```

```python
# import seaborn library
import seaborn as sns

# load iris dataset
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
# plot a line graph
iris.plot() 
```

```python
# plot an area graph
iris.plot(kind="area")
```

```python
# plot an area graph
iris.plot(kind="area", alpha=0.3)
```

```python
# plot an area graph
# plot with stacked = False
iris.plot(kind="area", alpha=0.3, stacked=False)
```

> **NOTE** stacked = True is default for area chart

## Scatter Plot
Its nothing plot of data points in the cartesian coordinate system where each axis may represent a colum of the dataset

```python
# plotting scatter plot for iris dataset where -
# x axis represent sepal length 
# y axis represent petal length
iris.plot.scatter(x = "sepal_length", y = "petal_length")
```

```python
# c axis represent sepal width
iris.plot.scatter(x = 'sepal_length', y = 'petal_length', c = 'sepal_width')
```

```python
ax = iris.plot.scatter(x = 'sepal_length', y = 'petal_length', label = 'Length');
iris.plot.scatter(x = 'sepal_width', y = 'petal_width', label = 'Width', ax = ax, color = 'r')
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'petal_length', c = 'sepal_width', s = iris['petal_width']*700)
```

```python
iris.plot.hexbin(x = 'sepal_length', y = 'petal_length', gridsize = 10, C = 'sepal_width')
```
